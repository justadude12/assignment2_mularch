```python
import matplotlib.pyplot as plt
```

```python
threads = [1, 2, 4, 8]
seq = [0.6849, 0.6941, 0.6818, 0.6945]
unopt = [1.073, 1.998, 1.896, 1.852]
opt = [0.6653, 0.4304, 0.2359, 0.09091]
```

```python
fig, ax = plt.subplots()
ax.plot(threads, seq, label="Sequential")
ax.plot(threads, unopt, label="Unoptimized")
ax.plot(threads, opt, label="Optimized")
ax.set_xlabel("Thread number")
ax.set_ylabel("Execution time [s]")
ax.legend()
```

```python
fig.savefig("2_1.png", dpi=300)
```
